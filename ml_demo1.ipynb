```python
# import neccesary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
# loading dataset and displaying
iris_data = pd.read_csv("/content/iris_data.csv")
```

```python
iris_data.head()
```

```python
# split the data into x and y
x = iris_data.drop(columns=['sepal_length','species'])
y = iris_data['species']
```

```python
x.head()
```

```python
# create ml model
model = LogisticRegression()
```

```python
# train the model
model.fit(x.values,y)
```

```python
# prediction using ml
predictions = model.predict([[3.5,4.0,2.3,0.2]])
```

```python
# print predictions
print(predictions)
```
